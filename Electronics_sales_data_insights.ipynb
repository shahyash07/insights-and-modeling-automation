```python
# Filter data to include only relevant columns
df_basket = df[['Order Date', 'Purchase Address', 'Product']]

# Group by Order Date and Purchase Address to find products bought together
df_basket = df_basket.groupby(['Order Date', 'Purchase Address'])['Product'].apply(list).reset_index()

# Generate product pairs bought together
product_combinations = []
for products in df_basket['Product']:
    if len(products) > 1:  # Only consider orders with multiple products
        product_combinations.extend(combinations(sorted(products), 2))

# Count occurrences of each pair
product_pair_counts = Counter(product_combinations)

# Convert to DataFrame for visualization
bought_together_df = pd.DataFrame(product_pair_counts.items(), columns=['Product Pair', 'Count'])
bought_together_df.sort_values(by='Count', ascending=False, inplace=True)

# Display top pairs
bought_together_df.head(10)
```

# Repeat customers analysis

```python
# i want to see how many customers are repating vs single time coming  -
#  1.4 M addresses and 1.8 M orders, order dates are more addreeses are less so there would berepat customers
# using purchase address to identify unique customers
# date
df['Order Day'] = pd.to_datetime(df['Order Date'], format='%m/%d/%y %H:%M').dt.date

# Count unique shopping days per customer
customer_visits = df.groupby('Purchase Address')['Order Day'].nunique().reset_index()
customer_visits.columns = ['Purchase Address', 'Unique Shopping Days']

# Determine single-visit vs repeat customers
customer_visits['Customer Type'] = customer_visits['Unique Shopping Days'].apply(lambda x: 'Single Visit' if x == 1 else 'Repeat Visit')

# Merge with original dataframe
df = df.merge(customer_visits, on='Purchase Address')
```

```python
df['Customer Type'].value_counts()
```

```python
# lets see how the customers repeats over months

# total customers per month
monthly_customers = df.groupby(['Month'], observed=False)['Purchase Address'].nunique().reset_index()
monthly_customers.columns = ['Month', 'Total Customers']

# Count repeat customers per month
repeat_customers = df[df['Customer Type'] == 'Repeat Visit'].groupby(['Month'], observed=False)['Purchase Address'].nunique().reset_index()
repeat_customers.columns = ['Month', 'Repeat Customers']

# Merge both counts and calculate Repeat Customer Rate
repeat_customers = monthly_customers.merge(repeat_customers, on='Month', how='left').fillna(0)
repeat_customers['Repeat Customer Rate'] = (repeat_customers['Repeat Customers'] / repeat_customers['Total Customers']) * 100

repeat_customers['Month'] = pd.Categorical(repeat_customers['Month'], categories=month_order, ordered=True)
repeat_customers = repeat_customers.sort_values('Month')
```
